## Uji Hipotesis

### Pisahkan data Pembelian antara Anggota Program Loyalitas dan Bukan Anggota
- Dengan data yang telah diolah kita akan memisahkan data pembelian antara anggota program loyalitas dan bukan anggota.

```python
# Pisahkan data pembelian antara anggota program loyalitas dan bukan anggota
loyal_customers = retail_cleaned[retail_cleaned['loyalty_program'] == 1]
non_loyal_customers = retail_cleaned[retail_cleaned['loyalty_program'] == 0]

# Cetak jumlah pembelian oleh anggota program loyalitas dan bukan anggota
print("Jumlah Pembelian oleh Anggota Program Loyalitas:", len(loyal_customers))
print("Jumlah Pembelian oleh Bukan Anggota Program Loyalitas:", len(non_loyal_customers))
```

```python
# Hitung jumlah pembelian oleh anggota program loyalitas dan bukan anggota
loyal_counts = len(loyal_customers)
non_loyal_counts = len(non_loyal_customers)

# Visualisasi perbandingan jumlah pembelian antara anggota program loyalitas dan bukan anggota menggunakan pie chart
plt.figure(figsize=(8, 8))
plt.pie([non_loyal_counts, loyal_counts], labels=['Non-Members', 'Members'], autopct='%1.1f%%', startangle=140)
plt.title('Perbandingan Loyalty Program Members dan Non-Members')
plt.axis('equal')
plt.show()
```

```python
loyalty_customers_mean_purchase = loyal_customers['quantity'].mean()
print("Rata-rata pembelian pelanggan anggota program loyalitas:", loyalty_customers_mean_purchase)
non_loyalty_customers_mean_purchase = non_loyal_customers['quantity'].mean()
print("Rata-rata pembelian pelanggan non-anggota program loyalitas:", non_loyalty_customers_mean_purchase)
```

<font color = green> Kesimpulan Sementara </font>
- Jumlah anggota program loyalitas memiliki jumlah yang lebih sedikit dibanding bukan anggota program loyalitas.

### Lakukan uji statistik, seperti uji t atau uji Mann-Whitney 
- Melakukan uji statistik, seperti uji t atau uji Mann-Whitney (bergantung pada distribusi data), untuk membandingkan rata-rata pembelian.

```python
# Pisahkan data pembelian antara anggota program loyalitas dan bukan anggota
loyal_customers = retail_cleaned[retail_cleaned['loyalty_program'] == 1]['quantity']
non_loyal_customers = retail_cleaned[retail_cleaned['loyalty_program'] == 0]['quantity']

# Lakukan uji t untuk membandingkan rata-rata pembelian
t_statistic, p_value = st.ttest_ind(loyal_customers, non_loyal_customers, equal_var=False)

# Tampilkan hasil uji statistik
print("Hasil Uji t (Data Terdistribusi Normal):")
print("Nilai t-statistic:", t_statistic)
print("Nilai p:", p_value)

# Ambil keputusan berdasarkan nilai p
alpha = 0.05
if p_value < alpha:
    print("Kesimpulan: Tolak H0, Ada perbedaan signifikan dalam rata-rata pembelian antara anggota program loyalitas dan bukan anggota.")
else:
    print("Kesimpulan: Tidak cukup bukti untuk menolak H0, Tidak ada perbedaan signifikan dalam rata-rata pembelian antara kedua kelompok.")
```

```python
# Lakukan uji Mann-Whitney untuk membandingkan rata-rata pembelian
u_statistic, p_value = st.mannwhitneyu(loyal_customers, non_loyal_customers)

# Tampilkan hasil uji statistik
print("Hasil Uji Mann-Whitney (Data Tidak Terdistribusi Normal):")
print("Nilai U-statistic:", u_statistic)
print("Nilai p:", p_value)

# Ambil keputusan berdasarkan nilai p
alpha = 0.05
if p_value < alpha:
    print("Kesimpulan: Tolak H0, Ada perbedaan signifikan dalam rata-rata pembelian antara anggota program loyalitas dan bukan anggota.")
else:
    print("Kesimpulan: Tidak cukup bukti untuk menolak H0, Tidak ada perbedaan signifikan dalam rata-rata pembelian antara kedua kelompok.")
```

<font color = green> Kesimpulan Sementara </font>
- Uji t
    - Karena nilai p (0.014) lebih kecil dari tingkat signifikansi yang umumnya digunakan (`biasanya α = 0.05`), kita memiliki cukup bukti untuk menolak hipotesis nol. **Artinya, kita dapat menyimpulkan bahwa ada perbedaan signifikan dalam rata-rata pembelian antara anggota program loyalitas dan bukan anggota**. Dengan kata lain, pelanggan yang menjadi anggota program loyalitas memiliki rata-rata pembelian yang berbeda secara signifikan dari pelanggan yang bukan anggota program loyalitas.
- Uji Mann-Whitney
    - Karena nilai p sangat kecil, kita memiliki cukup bukti untuk menolak hipotesis nol. **Artinya, kita dapat menyimpulkan bahwa ada perbedaan signifikan dalam rata-rata pembelian antara anggota program loyalitas dan bukan anggota**. Dengan kata lain, pelanggan yang menjadi anggota program loyalitas cenderung memiliki rata-rata pembelian yang berbeda secara signifikan dari pelanggan yang bukan anggota program loyalitas.

```python
# Memisahkan data pembelian antara anggota program loyalitas dan bukan anggota
purchases_loyal = retail_cleaned[retail_cleaned['loyalty_program'] == 1]['quantity']
purchases_non_loyal = retail_cleaned[retail_cleaned['loyalty_program'] == 0]['quantity']

# Melakukan uji t independen
t_statistic, p_value = ttest_ind(purchases_loyal, purchases_non_loyal)

# Menampilkan hasil uji statistik
print("Nilai t-statistic:", t_statistic)
print("Nilai p:", p_value)

# Menentukan keputusan
alpha = 0.05
if p_value < alpha:
    print("Kesimpulan: Tolak H0, Ada perbedaan signifikan dalam rata-rata pembelian antara anggota program loyalitas dan bukan anggota program loyalitas.")
else:
    print("Kesimpulan: Gagal menolak H0, Tidak terdapat perbedaan signifikan dalam rata-rata pembelian antara kedua kelompok.")
```
